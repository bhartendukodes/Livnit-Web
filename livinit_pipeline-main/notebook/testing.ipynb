```python
import os, re, json, time, html, requests, undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from supabase import create_client, Client

# ------------------------------------------------------------
# ENVIRONMENT VARIABLES
# ------------------------------------------------------------
SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
SUPABASE_BUCKET = os.getenv("SUPABASE_BUCKET", "3d-assets")
SUPABASE_TABLE = os.getenv("SUPABASE_TABLE", "assets")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
OPENAI_MODEL = os.getenv("OPENAI_MODEL", "gpt-4o-mini")

supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)

# ------------------------------------------------------------
# DRIVER
# ------------------------------------------------------------
def create_headless_driver():
    options = uc.ChromeOptions()
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("--window-size=1920,1080")
    driver = uc.Chrome(options=options)  # auto-detects Chrome version
    return driver


# ------------------------------------------------------------
# SCRAPER HELPERS
# ------------------------------------------------------------
def get_first_text(driver, xpaths):
    for xp in xpaths:
        try:
            el = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, xp)))
            txt = el.text.strip()
            if txt:
                return txt
        except Exception:
            pass
    return ""


def get_all_texts(driver, xpaths, limit=30):
    for xp in xpaths:
        try:
            els = driver.find_elements(By.XPATH, xp)
            out = [el.text.strip() for el in els[:limit] if el.text.strip()]
            if out:
                return out
        except Exception:
            pass
    return []


def get_first_html(driver, xpaths):
    for xp in xpaths:
        try:
            el = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, xp)))
            return el.get_attribute("outerHTML") or ""
        except Exception:
            pass
    return ""


def expand_product_details(driver):
    """Expands collapsed 'Product details' section if it exists."""
    try:
        button = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, "//button[@aria-label='Product details']"))
        )
        driver.execute_script("arguments[0].click();", button)
        time.sleep(2)
        print("üîΩ Expanded 'Product details' section")
    except Exception:
        pass


# ------------------------------------------------------------
# OPENAI HELPERS
# ------------------------------------------------------------
def extract_json_block(text):
    start = text.find("{")
    if start == -1:
        return None
    depth = 0
    for i in range(start, len(text)):
        if text[i] == "{":
            depth += 1
        elif text[i] == "}":
            depth -= 1
            if depth == 0:
                return text[start:i + 1]
    return None


def openai_chat_json(system_prompt, user_prompt):
    url = "https://api.openai.com/v1/chat/completions"
    headers = {"Authorization": f"Bearer {OPENAI_API_KEY}", "Content-Type": "application/json"}
    payload = {
        "model": OPENAI_MODEL,
        "temperature": 0.2,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    }
    resp = requests.post(url, headers=headers, json=payload, timeout=60)
    resp.raise_for_status()
    content = resp.json()["choices"][0]["message"]["content"]
    js = extract_json_block(content)
    return json.loads(js) if js else json.loads(content)


# ------------------------------------------------------------
# SUPABASE UPLOAD + CLEANUP
# ------------------------------------------------------------
def upload_to_supabase(metadata_url, data):
    prefix = f"{SUPABASE_URL}/storage/v1/object/public/{SUPABASE_BUCKET}/"
    rel_path = metadata_url.replace(prefix, "")
    encoded = json.dumps(data, indent=2).encode("utf-8")
    storage = supabase.storage.from_(SUPABASE_BUCKET)
    try:
        storage.remove([rel_path])  # ignore errors
    except Exception:
        pass
    storage.upload(rel_path, encoded)
    print(f"‚úÖ Uploaded metadata to {rel_path}")


# ------------------------------------------------------------
# MAIN LOGIC
# ------------------------------------------------------------
def test_single_product():
    # Option 1: pull one record from Supabase
    row = supabase.table(SUPABASE_TABLE).select("*").limit(1).execute().data[0]

    # Option 2: hardcode (uncomment if you want manual test)
    # row = {
    #     "name": "office_chairs_12",
    #     "category": "chairs",
    #     "product_url": "https://www.walmart.com/ip/Furmax-Office-Mid-Back-Swivel-Lumbar-Support-Desk-Computer-Ergonomic-Mesh-Chair-with-Armrest-Light-Green/566640994",
    #     "metadata_url": "https://nyvlydjdvhsunqbliqru.supabase.co/storage/v1/object/public/3d-assets/metadata/office_chairs_12.json"
    # }

    product_url = row["product_url"]
    metadata_url = row["metadata_url"]
    name = row.get("name", "asset")

    print(f"\n=== TESTING SINGLE PRODUCT ===")
    print(f"Name: {name}")
    print(f"URL:  {product_url}")

    # XPATH CONFIG
    TITLE_XPATHS = ["//h1", "//*[@data-automation-id='product-title']"]
    PRICE_XPATHS = ["//*[@data-automation-id='price']", "//span[contains(@class,'price')]"]
    DESCRIPTION_XPATHS = [
        "//*[@data-testid='product-description-content']",
        "//*[@data-automation-id='product-description']",
        "//div[contains(@class,'about-desc')]",
        "//div[h2[contains(., 'About this item')]]",
    ]
    BULLETS_XPATHS = ["//ul[li]//li", "//*[@data-automation-id='about-desc-contents']//li"]
    SPECS_XPATHS = ["//*[@data-automation-id='specifications']", "//div[h2[contains(., 'Specifications')]]", "//table"]

    driver = create_headless_driver()
    driver.get(product_url)
    time.sleep(8)

    expand_product_details(driver)

    title = get_first_text(driver, TITLE_XPATHS)
    price = get_first_text(driver, PRICE_XPATHS)
    bullets = get_all_texts(driver, BULLETS_XPATHS)
    desc_html = get_first_html(driver, DESCRIPTION_XPATHS)
    specs_html = get_first_html(driver, SPECS_XPATHS)
    body_text = driver.find_element(By.TAG_NAME, "body").text
    driver.quit()

    # ------------------------------------------------------------
    # LLM PROMPT
    # ------------------------------------------------------------
    user_prompt = f"""
URL: {product_url}
TITLE: {title}
PRICE: {price}
BULLETS: {'; '.join(bullets)}
DESCRIPTION_HTML: {desc_html}
SPECS_HTML: {specs_html}
PAGE_TEXT: {body_text[:10000]}
"""

    SCHEMA_GUIDANCE = """
You are given Walmart product text. Output ONLY a single JSON object with this structure:

{
  "assetMetadata": {"boundingBox": {"x": <m>, "y": <m>, "z": <m>}},
  "annotations": {
    "description": "<short>",
    "category": "<string>",
    "width": <cm>,
    "depth": <cm>,
    "height": <cm>,
    "volume": <cm3>,
    "materials": [],
    "composition": [],
    "mass": 0,
    "receptacle": false,
    "frontView": 0,
    "onCeiling": false,
    "onWall": false,
    "onFloor": true,
    "onObject": false,
    "uid": "<same as name>",
    "pose_z_rot_angle": 0.0,
    "scale": 1.0,
    "z_axis_scale": true
  }
}
Return ONLY JSON.
"""

    # ------------------------------------------------------------
    # RUN OPENAI EXTRACTION
    # ------------------------------------------------------------
    data = openai_chat_json(SCHEMA_GUIDANCE, user_prompt)
    data["annotations"]["uid"] = name

    predicted_category = data["annotations"].get("category", "").strip().lower()
    original_category = (row.get("category") or "").strip().lower()

    # ------------------------------------------------------------
    # CATEGORY CHECK
    # ------------------------------------------------------------
    if predicted_category and original_category and predicted_category != original_category:
        print(f"‚ö†Ô∏è Category mismatch: Supabase={original_category} ‚Üí Walmart={predicted_category}")
        prefix = f"{SUPABASE_URL}/storage/v1/object/public/{SUPABASE_BUCKET}/"
        rel_path = metadata_url.replace(prefix, "")
        supabase.table(SUPABASE_TABLE).delete().eq("name", name).execute()
        try:
            supabase.storage.from_(SUPABASE_BUCKET).remove([rel_path])
        except Exception:
            pass
        print(f"üóëÔ∏è Deleted {name} (category mismatch).")
        return

    # ------------------------------------------------------------
    # UPLOAD RESULT
    # ------------------------------------------------------------
    print("\nGenerated JSON preview:")
    print(json.dumps(data, indent=2)[:1000], "...\n")

    upload_to_supabase(metadata_url, data)
    print(f"‚úÖ Finished test for {name}")


# ------------------------------------------------------------
# RUN
# ------------------------------------------------------------
if __name__ == "__main__":
    test_single_product()
```

```python
# scrape_single_product_update.py
import os, re, json, time, requests, undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from supabase import create_client, Client
from dotenv import load_dotenv

# ------------------------------------------------------------
# ENVIRONMENT SETUP
# ------------------------------------------------------------
load_dotenv()
SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
SUPABASE_BUCKET = os.getenv("SUPABASE_BUCKET", "3d-assets")
SUPABASE_TABLE = os.getenv("SUPABASE_TABLE", "assets")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
OPENAI_MODEL = os.getenv("OPENAI_MODEL", "gpt-4o-mini")

supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)

# ------------------------------------------------------------
# DRIVER (HEADLESS)
# ------------------------------------------------------------
def create_headless_driver():
    options = uc.ChromeOptions()
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("--window-size=1920,1080")
    return uc.Chrome(options=options)  # auto-detect Chrome version


# ------------------------------------------------------------
# HTML HELPERS
# ------------------------------------------------------------
def get_first_text(driver, xpaths):
    for xp in xpaths:
        try:
            el = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, xp)))
            txt = el.text.strip()
            if txt:
                return txt
        except Exception:
            pass
    return ""


def get_all_texts(driver, xpaths, limit=30):
    for xp in xpaths:
        try:
            els = driver.find_elements(By.XPATH, xp)
            out = [el.text.strip() for el in els[:limit] if el.text.strip()]
            if out:
                return out
        except Exception:
            pass
    return []


def get_first_html(driver, xpaths):
    for xp in xpaths:
        try:
            el = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, xp)))
            return el.get_attribute("outerHTML") or ""
        except Exception:
            pass
    return ""


def expand_section(driver, label):
    try:
        btn = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, f"//button[@aria-label='{label}']"))
        )
        driver.execute_script("arguments[0].click();", btn)
        time.sleep(2)
        print(f"Expanded '{label}'")
    except Exception:
        pass


# ------------------------------------------------------------
# OPENAI HELPER
# ------------------------------------------------------------
def extract_json_block(text: str):
    start = text.find("{")
    if start == -1:
        return None
    depth = 0
    for i in range(start, len(text)):
        if text[i] == "{":
            depth += 1
        elif text[i] == "}":
            depth -= 1
            if depth == 0:
                return text[start:i+1]
    return None


def openai_chat_json(system_prompt: str, user_prompt: str):
    url = "https://api.openai.com/v1/chat/completions"
    headers = {"Authorization": f"Bearer {OPENAI_API_KEY}", "Content-Type": "application/json"}
    payload = {
        "model": OPENAI_MODEL,
        "temperature": 0.2,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
    }
    resp = requests.post(url, headers=headers, json=payload, timeout=60)
    resp.raise_for_status()
    content = resp.json()["choices"][0]["message"]["content"]
    js = extract_json_block(content)
    return json.loads(js) if js else json.loads(content)


# ------------------------------------------------------------
# SUPABASE UPLOAD (UPDATE ONLY)
# ------------------------------------------------------------
def upload_to_supabase_same_path(metadata_url: str, data: dict):
    prefix = f"{SUPABASE_URL}/storage/v1/object/public/{SUPABASE_BUCKET}/"
    rel_path = metadata_url.replace(prefix, "")
    encoded = json.dumps(data, indent=2).encode("utf-8")
    storage = supabase.storage.from_(SUPABASE_BUCKET)
    try:
        storage.remove([rel_path])
    except Exception:
        pass
    storage.upload(rel_path, encoded)
    print(f"‚úÖ Uploaded metadata to {rel_path}")


# ------------------------------------------------------------
# MAIN TEST FUNCTION
# ------------------------------------------------------------
def test_single_update_only():
    # Fetch one row from Supabase
    row = supabase.table(SUPABASE_TABLE).select("*").limit(1).execute().data[0]

    # (Optionally hardcode)
    # row = {
    #     "name": "accent_chair_13",
    #     "category": "accent_chair",
    #     "product_url": "https://www.walmart.com/ip/Mainstays-Boucle-Swivel-Accent-Chair-Cream/...",
    #     "metadata_url": "https://nyvlydjdvhsunqbliqru.supabase.co/storage/v1/object/public/3d-assets/metadata/accent_chair_13.json",
    # }

    name = row.get("name", "asset")
    category_supabase = (row.get("category") or "").strip().lower()
    product_url = row["product_url"]
    metadata_url = row["metadata_url"]

    print("\n=== TESTING SINGLE PRODUCT ===")
    print("Name:", name)
    print("URL:", product_url)

    TITLE_XPATHS = ["//h1", "//*[@data-automation-id='product-title']"]
    PRICE_XPATHS = ["//*[@data-automation-id='price']", "//span[contains(@class,'price')]"]
    DESCRIPTION_XPATHS = [
        "//*[@data-testid='product-description-content']",
        "//*[@data-automation-id='product-description']",
        "//div[contains(@class,'about-desc')]",
        "//div[h2[contains(., 'About this item')]]",
    ]
    BULLETS_XPATHS = ["//ul[li]//li", "//*[@data-automation-id='about-desc-contents']//li"]
    SPECS_XPATHS = [
        "//*[@data-automation-id='specifications']",
        "//div[h2[contains(., 'Specifications')]]",
        "//table",
    ]

    # ------------------------------------------------------------
    # SCRAPE PRODUCT PAGE
    # ------------------------------------------------------------
    driver = create_headless_driver()
    driver.get(product_url)
    time.sleep(8)

    expand_section(driver, "Product details")
    expand_section(driver, "More details")  # exposes Specifications if present

    title = get_first_text(driver, TITLE_XPATHS)
    price = get_first_text(driver, PRICE_XPATHS)
    bullets = get_all_texts(driver, BULLETS_XPATHS)
    desc_html = get_first_html(driver, DESCRIPTION_XPATHS)
    specs_html = get_first_html(driver, SPECS_XPATHS)
    body_text = driver.find_element(By.TAG_NAME, "body").text
    driver.quit()

    # ------------------------------------------------------------
    # OPENAI CALL
    # ------------------------------------------------------------
    schema_guidance = """
You are given Walmart product text and specifications.
Output ONLY one JSON object in SI units (meters):

{
  "assetMetadata": {"boundingBox": {"x": <m>, "y": <m>, "z": <m>}},
  "annotations": {
    "description": "<short>",
    "category": "<string>",
    "width": <m>,
    "depth": <m>,
    "height": <m>,
    "volume": <m3>,
    "materials": [],
    "composition": [],
    "mass": <kg>,
    "receptacle": <bool>,
    "frontView": 0,
    "onCeiling": false,
    "onWall": false,
    "onFloor": true,
    "onObject": false,
    "uid": "<name>",
    "pose_z_rot_angle": 0.0,
    "scale": 1.0,
    "z_axis_scale": true
  }
}
All lengths must be converted to METERS. Return ONLY JSON.
"""

    user_prompt = f"""
URL: {product_url}
TITLE: {title}
PRICE: {price}
BULLETS: {'; '.join(bullets)}
DESCRIPTION_HTML: {desc_html}
SPECS_HTML: {specs_html}
PAGE_TEXT: {body_text[:8000]}
"""

    data = openai_chat_json(schema_guidance, user_prompt)
    data.setdefault("annotations", {})
    data["annotations"]["uid"] = name

    # ------------------------------------------------------------
    # NORMALIZATION
    # ------------------------------------------------------------
    ann = data["annotations"]
    bbox = data["assetMetadata"].setdefault("boundingBox", {})

    # 1. Force numeric and meters
    for k in ("width", "depth", "height"):
        try:
            ann[k] = float(ann.get(k, 0))
        except Exception:
            ann[k] = 0.0

    # 2. Bounding box
    if not bbox or any(bbox.get(k, 0) == 0 for k in ("x", "y", "z")):
        bbox["x"], bbox["y"], bbox["z"] = ann["width"], ann["depth"], ann["height"]
        data["assetMetadata"]["boundingBox"] = {k: round(v, 4) for k, v in bbox.items()}

    # 3. Volume (m¬≥)
    if not ann.get("volume"):
        ann["volume"] = round(ann["width"] * ann["depth"] * ann["height"], 4)

    # 4. Composition
    if not ann.get("composition") and ann.get("materials"):
        n = len(ann["materials"])
        if n:
            vals = [round(1.0/n, 3) for _ in range(n)]
            vals[-1] = round(1.0 - sum(vals[:-1]), 3)
            ann["composition"] = vals

    # 5. Category map
    CATEGORY_MAP = {
        "furniture": "accent_chair",
        "chair": "accent_chair",
        "accent chair": "accent_chair",
        "sofa": "sofa",
        "couch": "sofa",
    }
    cat = (ann.get("category") or "").strip().lower()
    ann["category"] = CATEGORY_MAP.get(cat, cat)

    data["annotations"] = ann

    # ------------------------------------------------------------
    # PREVIEW + UPLOAD
    # ------------------------------------------------------------
    preview_path = f"_preview_{name}.json"
    with open(preview_path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)

    print("\n--- FINAL JSON (meters) ---")
    print(json.dumps(data, indent=2))
    print(f"\nSaved preview -> {preview_path}")

    upload_to_supabase_same_path(metadata_url, data)
    print("‚úÖ Done updating Supabase.")


# ------------------------------------------------------------
# RUN
# ------------------------------------------------------------
if __name__ == "__main__":
    test_single_update_only()
```

```python
import os, re, json, time, requests, undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from supabase import create_client, Client
from dotenv import load_dotenv

# ------------------------------------------------------------
# ENV
# ------------------------------------------------------------
load_dotenv()
SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
SUPABASE_BUCKET = os.getenv("SUPABASE_BUCKET", "3d-assets")
SUPABASE_TABLE = os.getenv("SUPABASE_TABLE", "assets")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
OPENAI_MODEL = os.getenv("OPENAI_MODEL", "gpt-4o-mini")

supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)

# ------------------------------------------------------------
# HEADLESS DRIVER
# ------------------------------------------------------------
def create_headless_driver():
    options = uc.ChromeOptions()
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("--window-size=1920,1080")
    return uc.Chrome(options=options)

# ------------------------------------------------------------
# HELPERS
# ------------------------------------------------------------
def expand_section(driver, label):
    try:
        btn = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, f"//button[@aria-label='{label}']"))
        )
        driver.execute_script("arguments[0].click();", btn)
        time.sleep(2)
        print(f"Expanded '{label}'")
    except Exception:
        pass

def get_first_text(driver, xpaths):
    for xp in xpaths:
        try:
            el = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, xp)))
            txt = el.text.strip()
            if txt:
                return txt
        except Exception:
            pass
    return ""

def get_all_texts(driver, xpaths, limit=30):
    for xp in xpaths:
        try:
            els = driver.find_elements(By.XPATH, xp)
            out = [el.text.strip() for el in els[:limit] if el.text.strip()]
            if out:
                return out
        except Exception:
            pass
    return []

def get_first_html(driver, xpaths):
    for xp in xpaths:
        try:
            el = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, xp)))
            return el.get_attribute("outerHTML") or ""
        except Exception:
            pass
    return ""

def extract_json_block(text: str):
    start = text.find("{")
    if start == -1:
        return None
    depth = 0
    for i in range(start, len(text)):
        if text[i] == "{":
            depth += 1
        elif text[i] == "}":
            depth -= 1
            if depth == 0:
                return text[start:i+1]
    return None

# ------------------------------------------------------------
# OPENAI CALL
# ------------------------------------------------------------
def openai_chat_json(system_prompt: str, user_prompt: str):
    url = "https://api.openai.com/v1/chat/completions"
    headers = {"Authorization": f"Bearer {OPENAI_API_KEY}", "Content-Type": "application/json"}
    payload = {
        "model": OPENAI_MODEL,
        "temperature": 0.2,
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
    }
    resp = requests.post(url, headers=headers, json=payload, timeout=70)
    if resp.status_code != 200:
        return None
    content = resp.json()["choices"][0]["message"]["content"]
    js = extract_json_block(content)
    try:
        return json.loads(js) if js else json.loads(content)
    except Exception:
        return None

# ------------------------------------------------------------
# FALLBACK PARSER (Regex for WxDxH)
# ------------------------------------------------------------
def try_parse_dimensions(text: str):
    patterns = [
        r"(\d+(?:\.\d+)?)\s*(?:in|inch|inches|\"|in\.)\s*[x√ó]\s*(\d+(?:\.\d+)?)\s*(?:in|inch|inches|\"|in\.)\s*[x√ó]\s*(\d+(?:\.\d+)?)\s*(?:in|inch|inches|\"|in\.)",
        r"(\d+(?:\.\d+)?)\s*W\s*[x√ó]\s*(\d+(?:\.\d+)?)\s*D\s*[x√ó]\s*(\d+(?:\.\d+)?)\s*H",
        r"(\d+(?:\.\d+)?)\s*cm\s*[x√ó]\s*(\d+(?:\.\d+)?)\s*cm\s*[x√ó]\s*(\d+(?:\.\d+)?)\s*cm",
    ]
    for pat in patterns:
        m = re.search(pat, text, flags=re.I)
        if m:
            vals = [float(x) for x in m.groups()]
            unit = "inches" if "in" in pat else "cm"
            if unit == "inches":
                vals = [v * 0.0254 for v in vals]  # convert to meters
            elif unit == "cm":
                vals = [v / 100.0 for v in vals]
            return tuple(vals)
    return (0.0, 0.0, 0.0)

# ------------------------------------------------------------
# SUPABASE UPLOAD (UPDATE)
# ------------------------------------------------------------
def upload_to_supabase_same_path(metadata_url: str, data: dict):
    prefix = f"{SUPABASE_URL}/storage/v1/object/public/{SUPABASE_BUCKET}/"
    rel_path = metadata_url.replace(prefix, "")
    encoded = json.dumps(data, indent=2).encode("utf-8")
    storage = supabase.storage.from_(SUPABASE_BUCKET)
    try:
        storage.remove([rel_path])
    except Exception:
        pass
    storage.upload(rel_path, encoded)
    print(f"‚úÖ Updated {rel_path}")

# ------------------------------------------------------------
# MAIN LOOP
# ------------------------------------------------------------
def process_all_products():
    all_rows = supabase.table(SUPABASE_TABLE).select("*").execute().data
    fail_streak = 0

    print(f"\n=== STARTING FULL RUN ({len(all_rows)} products) ===\n")

    for i, row in enumerate(all_rows, start=1):
        name = row.get("name", f"asset_{i}")
        product_url = row.get("product_url")
        metadata_url = row.get("metadata_url")
        category_supabase = (row.get("category") or "").lower().strip()

        print(f"\n{'='*70}")
        print(f"[{i}/{len(all_rows)}] {name}")
        print(product_url)

        driver = create_headless_driver()
        try:
            driver.get(product_url)
            time.sleep(8)
            expand_section(driver, "Product details")
            expand_section(driver, "More details")

            title = get_first_text(driver, ["//h1", "//*[@data-automation-id='product-title']"])
            price = get_first_text(driver, ["//*[@data-automation-id='price']", "//span[contains(@class,'price')]"])
            bullets = get_all_texts(driver, ["//ul[li]//li"])
            desc_html = get_first_html(driver, [
                "//*[@data-testid='product-description-content']",
                "//*[@data-automation-id='product-description']",
                "//div[contains(@class,'about-desc')]"
            ])
            specs_html = get_first_html(driver, [
                "//*[@data-automation-id='specifications']",
                "//div[h2[contains(., 'Specifications')]]",
                "//table"
            ])
            body_text = driver.find_element(By.TAG_NAME, "body").text
        except Exception as e:
            print(f"‚ùå Error loading product: {e}")
            driver.quit()
            fail_streak += 1
            if fail_streak >= 2:
                print("üõë Stopping run ‚Äî consecutive failures limit reached.")
                return
            continue

        driver.quit()

        schema_guidance = """
Output ONLY JSON with all metric units (meters):

{
  "assetMetadata": {"boundingBox": {"x": <m>, "y": <m>, "z": <m>}},
  "annotations": {
    "description": "<short>",
    "category": "<string>",
    "width": <m>,
    "depth": <m>,
    "height": <m>,
    "volume": <m3>,
    "materials": [],
    "composition": [],
    "mass": <kg>,
    "receptacle": <bool>,
    "frontView": 0,
    "onCeiling": false,
    "onWall": false,
    "onFloor": true,
    "onObject": false,
    "uid": "<name>",
    "pose_z_rot_angle": 0.0,
    "scale": 1.0,
    "z_axis_scale": true
  }
}
"""

        user_prompt = f"""
TITLE: {title}
PRICE: {price}
BULLETS: {'; '.join(bullets)}
DESCRIPTION_HTML: {desc_html}
SPECS_HTML: {specs_html}
PAGE_TEXT: {body_text[:8000]}
"""

        data = openai_chat_json(schema_guidance, user_prompt)

        if not data:
            fail_streak += 1
            print(f"‚ö†Ô∏è OpenAI returned no data ({fail_streak} consecutive fails).")
            if fail_streak >= 2:
                print("üõë Stopping run ‚Äî 2 consecutive API failures.")
                return
            # fallback with regex
            print("üß© Using fallback dimension parser...")
            w, d, h = try_parse_dimensions(body_text)
            data = {
                "assetMetadata": {"boundingBox": {"x": w, "y": d, "z": h}},
                "annotations": {
                    "description": title,
                    "category": category_supabase,
                    "width": w,
                    "depth": d,
                    "height": h,
                    "volume": round(w*d*h, 4),
                    "materials": [],
                    "composition": [],
                    "mass": 0,
                    "receptacle": False,
                    "frontView": 0,
                    "onCeiling": False,
                    "onWall": False,
                    "onFloor": True,
                    "onObject": False,
                    "uid": name,
                    "pose_z_rot_angle": 0.0,
                    "scale": 1.0,
                    "z_axis_scale": True
                }
            }
        else:
            fail_streak = 0

        # normalize numbers + fill fields
        ann = data["annotations"]
        bbox = data["assetMetadata"].setdefault("boundingBox", {})

        for k in ("width", "depth", "height"):
            try:
                ann[k] = float(ann.get(k, 0))
            except Exception:
                ann[k] = 0.0

        if not bbox or any(bbox.get(k, 0) == 0 for k in ("x", "y", "z")):
            bbox["x"], bbox["y"], bbox["z"] = ann["width"], ann["depth"], ann["height"]

        if not ann.get("volume"):
            ann["volume"] = round(ann["width"] * ann["depth"] * ann["height"], 4)

        if not ann.get("composition") and ann.get("materials"):
            n = len(ann["materials"])
            if n:
                vals = [round(1.0/n, 3) for _ in range(n)]
                vals[-1] = round(1.0 - sum(vals[:-1]), 3)
                ann["composition"] = vals

        data["annotations"] = ann
        data["assetMetadata"]["boundingBox"] = {k: round(v, 4) for k, v in bbox.items()}

        preview_path = f"_preview_{name}.json"
        with open(preview_path, "w", encoding="utf-8") as f:
            json.dump(data, f, indent=2, ensure_ascii=False)

        print(f"‚úÖ JSON ready for {name} -> {preview_path}")
        upload_to_supabase_same_path(metadata_url, data)
        print(f"‚úÖ Finished {name}")

    print("\nüéâ ALL PRODUCTS PROCESSED SUCCESSFULLY üéâ")


# ------------------------------------------------------------
# RUN
# ------------------------------------------------------------
if __name__ == "__main__":
    process_all_products()
```
